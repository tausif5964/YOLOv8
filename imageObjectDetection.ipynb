```python
# Install ultralytics
!pip install ultralytics

from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import glob

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Upload an image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Run inference (YOLO automatically saves in runs/detect/predict/)
results = model.predict(image_path, save=True, conf=0.5)

# Find the saved image path (YOLO saves with same filename inside predict folder)
saved_images = glob.glob("runs/detect/predict/*.jpg")  # look for jpgs in predict folder
output_path = saved_images[0]  # take the first result

# Load and display the annotated image
img = cv2.imread(output_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis("off")
plt.show()
```
