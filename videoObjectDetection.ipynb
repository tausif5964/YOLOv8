```python
# Install ultralytics
!pip install ultralytics

from ultralytics import YOLO
from google.colab import files
import glob
import os

# Upload a video
uploaded = files.upload()
video_path = list(uploaded.keys())[0]

# Load YOLOv8 model
model = YOLO("yolov8n.pt")  # Nano version for speed

# Run detection on video (save=True automatically saves annotated video)
results = model.predict(source=video_path, conf=0.5, save=True)

# Grab the saved video path
output_dir = results[0].save_dir  # Folder where YOLO saved results
output_videos = glob.glob(os.path.join(output_dir, "*.mp4"))

if len(output_videos) == 0:
    print("No annotated video found!")
else:
    output_video = output_videos[0]
    print("Annotated video saved at:", output_video)

    # Download the annotated video
    files.download(output_video)
```
