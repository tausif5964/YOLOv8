```python
# ==============================
# Step 2: Mount Google Drive
# ==============================
from google.colab import drive
drive.mount('/content/drive')

# ==============================
# Step 3: Import Libraries
# ==============================
from ultralytics import YOLO
from google.colab import files
import cv2
import matplotlib.pyplot as plt
```

```python
# ==============================
# Step 4: Load Fine-tuned Model from Drive
# ==============================
# Replace with your model path inside Google Drive
model_path = "/content/drive/MyDrive/YOLOv8_dental/best.pt"
model = YOLO(model_path)

# ==============================
# Step 5: Upload Test Image from Device
# ==============================
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # get uploaded filename

# ==============================
# Step 6: Run Inference on Image
# ==============================
results = model.predict(source=image_path, conf=0.5, show=True)

# ==============================
# Step 7: Save and Display Result
# ==============================
for r in results:
    r.save(filename="output.jpg")

# Display with matplotlib
img = cv2.imread("output.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```
